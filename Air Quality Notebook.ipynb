```python
#Numerical features vs target visualization 
def numeric_visualization(column):
    fig, ax = plt.subplots(1, 2, figsize=(15, 5))
    colors = {'normal': tail, 'abnormal': orange}

    #Histogram
    sns.histplot(x=column, data=df, hue=df['label'], palette=colors, ax=ax[0], bins=20)
    ax[0].set_title(f'{column} Histogram', fontsize=18)
    ax[0].tick_params(axis='both', which='major', labelsize=15)
    ax[0].set_xlabel(column, fontsize=18)
    ax[0].set_ylabel('Count', fontsize=18)

    #Box plot
    sns.boxplot(x=column, data=df, y=df['label'], ax=ax[1], palette=colors)
    ax[1].set_title(f'{column} Box Plot', fontsize=18)
    ax[1].tick_params(axis='both', which='major', labelsize=15)
    ax[1].set_xlabel(column, fontsize=18)
    ax[1].set_ylabel('Label', fontsize=18)

    plt.tight_layout()
    plt.show()
```

```python
#Comparison of statistical measures in different classes
def feature_statistics(column):
    #Extract columns
    normal = df[df['label']=='normal'][column]
    abnormal = df[df['label']=='abnormal'][column]


    #normal class
    normal_mean = round(normal.mean(), 2)
    normal_median = round(normal.median(), 2)
    normal_std = round(normal.std(), 2)
    normal_min = round(normal.min(), 2)
    normal_max = round(normal.max(), 2)
    normal_iqr = round(normal.quantile(0.75)-normal.quantile(0.25), 2)

    #abnormal class
    abnormal_mean = round(abnormal.mean(), 2)
    abnormal_median = round(abnormal.median(), 2)
    abnormal_std = round(abnormal.std(), 2)
    abnormal_min = round(abnormal.min(), 2)
    abnormal_max = round(abnormal.max(), 2)
    abnormal_iqr = round(abnormal.quantile(0.75)-abnormal.quantile(0.25), 2)



    result = pd.DataFrame({'Measure': ['Mean', 'Median', 'Std Deviation', 'Min', 'Max', 'IQR'],
                          'normal' : [normal_mean, normal_median, normal_std, normal_min,normal_max, normal_iqr],
                          'abnormal': [abnormal_mean, abnormal_median, abnormal_std, abnormal_min, abnormal_max, abnormal_iqr]}).set_index('Measure')


    return result
```

```python
#Categorical features visualization 
def category_visualization(column):
    plt.figure(figsize=(10, 6))
    df_counts = df[column].value_counts(normalize=True).reset_index()
    df_counts.columns = [column, 'percentage']
    sns.barplot(x=column, y='percentage', data=df_counts, palette=palette, edgecolor='black', order=df_counts[column])  
    plt.xticks(rotation=90)
    plt.ylabel('Percentage')
    plt.title(f'{column} Distribution', fontsize=13)
```

```python
#Categorical features vs target visualization 
def category_label_visualization(column):
    colors = {'normal': tail, 'abnormal': orange}  
    df_counts = df.groupby([column, 'label']).size().reset_index(name='count')
    pivot_df = df_counts.pivot(index=column, columns='label', values='count').fillna(0)

    pivot_df_percentage = pivot_df.div(pivot_df.sum(axis=1), axis=0)

    ax = pivot_df_percentage.plot(kind='bar', stacked=True, figsize=(10, 6), color=[colors.get(x) for x in pivot_df_percentage.columns], alpha=0.7, edgecolor='black')

    plt.ylabel('Percentage')
    plt.title(f'Percentage Distribution of {column} by Label')
    plt.xticks(rotation=90)
    plt.legend(title='Label', loc='upper right')

    plt.show()
```

```python
#Count of each category in features
def category_count(column):
    category_counts = df[column].value_counts()
    category_percentage = round(df[column].value_counts(normalize=True), 4)
    
    result = pd.DataFrame({
        'Category': category_counts.index,
        'Count': category_counts.values,
        'Percentage': category_percentage.values}).set_index('Category')
    
    return result
```

```python
#Numerical features
numerical_columns=df.select_dtypes(include=['float64', 'int64']).columns 

for col in numerical_columns:
    iqr = df[col].quantile(0.75) - df[col].quantile(0.25)
    lower_bound = df[col].quantile(0.25) - 1.5 * iqr
    upper_bound = df[col].quantile(0.75) + 1.5 * iqr
    df[col] = np.clip(df[col], lower_bound, upper_bound)
```

```python
#Categorical features
categorical_columns = ['boxName', 'Time of Day', 'Peak/NoPeak', 'Week Day', 'Weather'] 

threshold = 0.01
for col in categorical_columns:
    value_counts = df[col].value_counts()
    for category, count in value_counts.items():
        if count / df.shape[0] < threshold:
            df[col] = df[col].replace(category, df[col].mode()[0])
```

<div style="text-align: left; padding: 20px;">
    <span style="color: #00847E; font-size: 25px; font-weight: bold;">sin/cos Transformation
</span>
</div>

When dealing with cyclical (periodic) data, such as hours, months, or days of the week, using raw numerical values can mislead the model. 
Sin/Cos transformation helps maintain the cyclical relationship, ensuring that the model understands that the end of the cycle is close to the beginning.

```python
#Apply the transformation
periodic_cols = {
    'hour' : 23,
    'minute' : 59,
    'second' : 59,
    'month day' : 31,
    'month' : 12
}

for col, max_value in periodic_cols.items():
    df[f'{col}_sin'] = np.sin(2 * np.pi * df[col] / max_value)
    df[f'{col}_cos'] = np.cos(2 * np.pi * df[col] / max_value)
```

```python
#Function to evaluate model with custom threshold
def model_evaluation(name, classifier, params, X_train=X_train, X_test=X_test, y_train=y_train, y_test=y_test, threshold=0.5):
    cv = GridSearchCV(classifier, params, cv=3)
    cv.fit(X_train, y_train)
    
    #Get probability predictions
    y_prob = cv.predict_proba(X_test)  
    
    #Apply custom threshold
    y_pred = (y_prob[:, 1] > threshold).astype(int)
    

    #Metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    
    print(f'Classifier {name}')
    print(f'Best Parameters: {cv.best_params_}')
    print(f'Threshold: {threshold}')
    print(f'Accuracy: {accuracy}')
    print(f'Precision: {precision}')
    print(f'Recall: {recall}')
    print(f'F1 score: {f1}')
    print(classification_report(y_test, y_pred))
    
    
    #Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='mako', xticklabels=['Pred 0', 'Pred 1'], yticklabels=['Actual 0', 'Actual 1'])
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title('Confusion Matrix')
    plt.show()
    
    
    #Normalized Confusion Matrix
    cm_normalized = confusion_matrix(y_test, y_pred, normalize='true')
    plt.figure(figsize=(6, 6))
    sns.heatmap(cm_normalized, annot=True, fmt='.2f', cmap='mako', xticklabels=['Pred 0', 'Pred 1'], yticklabels=['Actual 0', 'Actual 1'])
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title('Normalized Confusion Matrix')
    plt.show()
    
    
    #ROC-AUC Curve for both classes
    plt.figure(figsize=(6, 6))
    for i, color in zip([0, 1], [orange, tail]):
        fpr, tpr, _ = roc_curve(y_test, y_prob[:, i], pos_label=i)
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, color=color, lw=2, label=f'Class {i} (area = {roc_auc:.2f})')

    plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC-AUC Curve')
    plt.legend()
    plt.show()

    
    #Precision-Recall Curve for both classes
    plt.figure(figsize=(6, 6))
    for i, color in zip([0, 1], [orange, tail]):
        precision_vals, recall_vals, _ = precision_recall_curve(y_test, y_prob[:, i], pos_label=i)
        plt.plot(recall_vals, precision_vals, color=color, lw=2, label=f'Class {i}')

    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('Precision-Recall Curve')
    plt.legend()
    plt.show()
```
